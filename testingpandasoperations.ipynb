```python
df['Join_Dates'] = pd.to_datetime(df['Join_Dates'])

```

```python
today = datetime.today()
```

```python
today
```

```python
df['Years_with_Company'] = df['Join_Dates'].apply(lambda x: today.year - x.year - ((today.month, today.day) < (x.month, x.day)))
```

```python
df_sorted = df.sort_values(by='Years_with_Company', ascending=False)
```
